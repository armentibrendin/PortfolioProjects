```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjust the config of the plots created
```

```python
# read in the movie data

df = pd.read_csv("/Users/brendinarmenti/Downloads/movies.csv")

#check out top five movies to make sure data is ok
df.head()
```

```python
# Going to be looking at correlations
```

```python
#Check for any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Extract rows where there is data present in all columns of dataframe.
clean_df = df.dropna()
clean_df
```

```python
#confirm the new dataframe is cleaned of any rows that have missing data
for col in clean_df.columns:
    pct_missing = np.mean(clean_df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Basic data cleaning

#Data types for columns
df.dtypes
```

```python
# Remove trailing decimal on budget, gross, votes columns
clean_df["budget"] = clean_df["budget"].astype('int64')
clean_df["gross"] = clean_df["gross"].astype('int64')
clean_df["votes"] = clean_df["votes"].astype('int64')
```

```python
clean_df
```

```python
# because the year is mentioned in two column (released and year), creat a new column (take first four values of released)
#turn into string to pull from it
# r'(\d{4})' grabs the first 4 digits in the string

clean_df["Correct Year"] = clean_df["released"].astype(str).str.extract(r'(\d{4})')
clean_df
```

```python
# order data by gross revenue in descending order
clean_df = clean_df.sort_values(by=['gross'], inplace=False, ascending=False)
clean_df
```

```python
# see more of the data

#all rows are being displayed instead of a truncated view
pd.set_option('display.max_rows', 20)
```

```python
#check for duplicates
clean_df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
clean_df
```

```python
#What columns are most correlated to the gross values for each movie?

#In theory, higher gross should mean higher budget(high correlation)
#The company should also be a good indicator of the budget(warner bros and other big companies can spend a lot)
```

```python
#Scatter plot comparing budget and gross revenue
plt.scatter(x=clean_df['budget'], y=clean_df['gross'])

plt.title("Budget vs Gross Earnings")
plt.xlabel("Budget for Film")
plt.ylabel("Gross Earnings")
plt.show()
```

```python
clean_df.head()
```

```python
#Plot budget vs gross

sns.regplot(x='budget', y='gross', data=clean_df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
plt.show()
```

```python
#Correlation values between columns numerically

# High correlation between budget and gross is confirmed
clean_df.corr(method='pearson', numeric_only=True)
```

```python
correlation_matrix = clean_df.corr(method='pearson', numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")

plt.show()
```

```python
# Look at company correlations (non numeric)

clean_df.head()
```

```python
#Giving the names of the companies a rank so calulations can be made

df_numerized = clean_df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
#All columns in the dataframe have been correlated now that the columns are are calculable

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs["gross"]
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
#filter out pairs that are higher than .5 correlation but also less that .98 to filter out intersection matrices
high_correlation = sorted_pairs[(sorted_pairs > 0.5) & (sorted_pairs < .98)]
high_correlation.drop_duplicates()
```
